```python
import pandas as pd

from tableone import TableOne

# !pip install tableone
```

```python
rhc = pd.read_csv('./rhc.csv').drop(['Unnamed: 0'], axis=1)

rhc.columns
```

```python
condition = {
    'arf': 'ARF',
    'chf': 'CHF',
    'cirr': 'Cirrhosis',
    'colcan': 'Colon Cancer',
    'coma': 'Coma',
    'copd': 'COPD',
    'lungcan': 'Lung Cancer',
    'mosf': 'MOSF w/Malignancy',
    'sepsis': 'MOSF w/Sepsis'
}

for column, value in list(zip(condition.keys(), condition.values())):
    rhc[column] = rhc['cat1'] == value

rhc['female'] = rhc['sex'] == 'Female'
rhc['died'] = rhc['death'] == 'Yes'
rhc['treatment'] = rhc['swang1'] == 'RHC'

# covariates we will use
covariables = ['female', 'age', 'meanbp1'] + list(condition.keys())
outcome = 'died'
treatment = 'treatment'

variables = covariables + [outcome, treatment]

```

```python
table1 = TableOne(rhc, columns=covariables, groupby=treatment, pval=True, htest_name=True, smd=True)
table1
```
